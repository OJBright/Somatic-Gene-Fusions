# STEP 1: Input Sequences and clean

```python
dna_fs = input("Enter the fusion sequence: ")

gene_a_cds = input("Enter Gene A CDS sequence: ")

gene_b_cds = input("Enter Gene B CDS sequence: ")

gene_a = gene_a_cds.replace(" ", "").upper()
gene_b = gene_b_cds.replace(" ", "").upper()
dna_fs = dna_fs.replace(" ", "").upper()
```

 # Step 2: Split the fusion sequence and search for each fragment

```python
left_fragment, right_fragment = dna_fs.split("*")

def search(fragment, gene):

    position = gene.find(fragment)

    if position == -1:
        return "Not found"

    return position


print("\nLEFT FRAGMENT")
print("Gene A:", search(left_fragment, gene_a))
print("Gene B:", search(left_fragment, gene_b))

print("\nRIGHT FRAGMENT")
print("Gene A:", search(right_fragment, gene_a))
print("Gene B:", search(right_fragment, gene_b))


left_in_a = gene_a.find(left_fragment)
left_in_b = gene_b.find(left_fragment)

right_in_a = gene_a.find(right_fragment)
right_in_b = gene_b.find(right_fragment)
```

# Step 4: Reconstruct Fusion CDS, save Junction

```python
if left_in_a != -1 and right_in_b != -1:

    left_sequence = gene_a[:left_in_a + len(left_fragment)]
    right_sequence = gene_b[right_in_b:]

elif left_in_b != -1 and right_in_a != -1:

    left_sequence = gene_b[:left_in_b + len(left_fragment)]
    right_sequence = gene_a[right_in_a:]

else:

    print("\nUnable to reconstruct the fusion CDS.")
    exit()


fusion_cds = left_sequence + right_sequence

print("\nFusion CDS:\n")
print(fusion_cds)

junction = len(left_sequence)

print("\nDNA junction position:", junction)
print("Reading frame:", junction % 3)
```

# Step 5: Translate Fusion CDS

```python
genetic_code = {

    "ATA":"I","ATC":"I","ATT":"I","ATG":"M",
    "ACA":"T","ACC":"T","ACG":"T","ACT":"T",
    "AAC":"N","AAT":"N","AAA":"K","AAG":"K",
    "AGC":"S","AGT":"S","AGA":"R","AGG":"R",
    "CTA":"L","CTC":"L","CTG":"L","CTT":"L",
    "CCA":"P","CCC":"P","CCG":"P","CCT":"P",
    "CAC":"H","CAT":"H","CAA":"Q","CAG":"Q",
    "CGA":"R","CGC":"R","CGG":"R","CGT":"R",
    "GTA":"V","GTC":"V","GTG":"V","GTT":"V",
    "GCA":"A","GCC":"A","GCG":"A","GCT":"A",
    "GAC":"D","GAT":"D","GAA":"E","GAG":"E",
    "GGA":"G","GGC":"G","GGG":"G","GGT":"G",
    "TCA":"S","TCC":"S","TCG":"S","TCT":"S",
    "TTC":"F","TTT":"F","TTA":"L","TTG":"L",
    "TAC":"Y","TAT":"Y","TAA":"*","TAG":"*","TGA":"*"

}

protein = ""

for i in range(0, len(fusion_cds)-2, 3):

    codon = fusion_cds[i:i+3]

    protein += genetic_code.get(codon, "X")


junction_aa = junction // 3

print("\nFusion protein sequence:\n")
print(protein)

print("\nProtein junction position:", junction_aa)

if junction_aa < len(protein):
    print("Amino acid at junction:", protein[junction_aa])
```

# Step 6: Generate every petide

```python
window_size = int(input("\nEnter peptide length: "))

all_peptides = []

for i in range(len(protein) - window_size + 1):

    peptide = protein[i:i + window_size]

    all_peptides.append((i, peptide))


print("\nALL PEPTIDES\n")

for position, peptide in all_peptides:

    print(position, peptide)
```

# Step 9: Identify Peptides  crossing Fusion junction

```python
fusion_peptides = []

for position, peptide in all_peptides:

    if position <= junction_aa < position + window_size:

        fusion_peptides.append((position, peptide))


print("\nFUSION-SPANNING PEPTIDES\n")

for position, peptide in fusion_peptides:

    print(position, peptide)
```

```python
import pandas as pd
from Bio import Entrez, SeqIO

# Enter email 
Entrez.email = "ojeromebright@gmail.com"

# Read TSV

tsv_file = input("Enter path to fusion TSV: ")

df = pd.read_csv(tsv_file, sep="\t")

print("Rows in file:", len(df))

# Keep only in-frame fusions

df = df[
    (df["frequency"] >= 3) &
    (df["Predicted_effect"] == "in-frame")
]

print("Rows after filtering:", len(df))

# Function to download CDS from NCBI

def get_cds(gene):

    print("Retrieving CDS for", gene)

    # Search Gene database
    search = Entrez.esearch(
        db="gene",
        term=f"{gene}[Gene Name] AND Homo sapiens[Organism]"
    )

    search_record = Entrez.read(search)

    if len(search_record["IdList"]) == 0:
        return None

    gene_id = search_record["IdList"][0]

    # Retrieve Gene record
    handle = Entrez.efetch(
        db="gene",
        id=gene_id,
        rettype="xml"
    )

    records = Entrez.read(handle)

    gene_record = records[0]

    accession = None

    # Search for an mRNA accession
    for locus in gene_record["Entrezgene_locus"]:

        try:

            accession = locus["Gene-commentary_accession"]

            if accession.startswith("NM_"):
                break

        except:
            pass

    if accession is None:
        return None

    print("Using transcript:", accession)

    # Download nucleotide sequence

    handle = Entrez.efetch(
        db="nucleotide",
        id=accession,
        rettype="fasta_cds_na",
        retmode="text"
    )

    record = SeqIO.read(handle, "fasta")

    cds = str(record.seq).upper()

    return cds


# Process each fusion

for index, row in df.iterrows():

    print("Fusion", index + 1)

    # Gene pair
    gene_a, gene_b = row["canonical_pair"].split("|")

    print("Gene A:", gene_a)
    print("Gene B:", gene_b)

    # Retrieve CDS

    gene_a_cds = get_cds(gene_a)

    gene_b_cds = get_cds(gene_b)

    if gene_a_cds is None or gene_b_cds is None:

        print("Unable to retrieve one or both CDSs.")
        continue

    # Multiple fusion sequences

    fusion_sequences = row["fusion_sequence"].split(";")

    print("Number of breakpoint sequences:",
          len(fusion_sequences))

    # Process each breakpoint independently

    for fusion in fusion_sequences:

        fusion = fusion.strip().upper()

        print()

        print("Breakpoint sequence")

        print(fusion)

        # Part 2 starts here...
```
